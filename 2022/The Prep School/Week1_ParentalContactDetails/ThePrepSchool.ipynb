```python
# import pandas
import pandas as pd
```

```python
# import dataset
df = pd.read_csv("PD 2022 Wk 1 Input - Input.csv", index_col=0)
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
df.head(5)
```

```python
# Pupil's name
df["Pupil's Name"] = df[["pupil last name","pupil first name"]].agg(', '.join, axis=1)
df.head(5)
```

```python
# Function to get full name of the parent to be contacted
def parental_full_name(row):
    parent = row[f'Parental Contact Name_{row["Parental Contact"]}']
    return f'{row["pupil last name"]}, {parent}'

# Generating Parent Full Name Column
df['Parental Contact Full Name'] = df.apply(parental_full_name, axis=1)
```

```python
# Function to get the email of the parent to be contacted
def parental_email(row):
    parent = row[f'Parental Contact Name_{row["Parental Contact"]}']
    return f'{parent}.{row["pupil last name"]}@{row["Preferred Contact Employer"]}.com'

# Generating Parent Email Address Column
df['Parental Contact Email Address'] = df.apply(parental_email,axis=1)
```

```python
# Function to get academic year based on date of birth
year_1 = pd.to_datetime('2014-09-01')
year_2 = pd.to_datetime('2013-08-31')
year_3 = pd.to_datetime('2012-08-31')
year_4 = pd.to_datetime('2011-08-31')

def academic_year(row):
    if row['Date of Birth'] > year_1:
        return 1
    elif row['Date of Birth'] > year_2:
        return 2
    elif row['Date of Birth'] > year_3:
        return 3
    else:
        return 4
    
# Generating Academic Year Column 
df['Academic Year'] = df.apply(academic_year,axis=1)
df.head(5)
```

```python
# Filtering to only include necessary columns
df_columns = ["Academic Year", "Pupil's Name", "Parental Contact Full Name", "Parental Contact Email Address"]
df = df[df_columns]
```

```python
# Setting Academic Year as the index
df.set_index('Academic Year', inplace=True)
df
```
