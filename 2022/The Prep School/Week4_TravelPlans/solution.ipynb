```python
# Import pandas
import pandas as pd
```

```python
# Import student dataset
df_1 = pd.read_csv("PD 2022 Wk 1 Input - Input.csv", index_col=0)
df_1.head(5)
```

```python
# Import method of transportation dataset
df_2 = pd.read_csv("PD 2021 WK 1 to 4 ideas - Preferences of Travel.csv")
df_2.head(5)
```

```python
# Reshaping the dataframe using melt to transform wide data into long format  
df = pd.melt(df_2, id_vars='Student ID', var_name='Weekday', value_name='Method of Travel')
df.set_index('Student ID', inplace=True)
df.sort_index()
```

```python
# Listing unique values in the 'Method of Travel' column using unique
df['Method of Travel'].unique()
```

```python
# Correcting typos in the data
df.replace(['Walkk', 'Waalk', 'Wallk', 'WAlk'], 'Walk', inplace=True)
df.replace(['Scoter', 'Scootr'], 'Scooter', inplace=True)
df.replace('Bycycle', 'Bicycle', inplace=True)
df.replace('Helicopeter', 'Helicopter', inplace=True)
df.replace('Carr', 'Car', inplace=True)

# Verifying the corrected data for typos
df['Method of Travel'].unique()
```

```python
# Grouping the DataFrame by 'Method of Travel' and 'Weekday' and calculaating the size of each group using size() 
df = df.groupby(by=['Method of Travel', 'Weekday'], axis=0, as_index=False).size()

# Set the name of the generated size column to 'Number of Trips'
df.rename(columns={'size':'Number of Trips'}, inplace=True)
df.head(5)
```

```python
# Adding a new column 'Trips per day' with a default value of 1000 (1000 students)
df['Trips per day'] = 1000
df.head(5)
```

```python
# Function to calculate the percentage of trips per day for each method of travel
def percent_trips(row):
    return row['Number of Trips']/row['Trips per day']

# Adding a new column '% of trips per day' by applying the function 'percent_trips'
df['% of trips per day'] = df.apply(percent_trips, axis=1)

# Applying round() to the '% of trips per day' column to round the values to the 2nd decimal place
df['% of trips per day'] = df['% of trips per day'].round(2)
df.head(5)
```

```python
# List of sustainable methods of travel
sustainable = ['Bicycle', "Dad's Shoulders", 'Hopped', 'Jumped', "Mum's Shoulders", 'Scooter', 'Skipped', 'Walk']

# Function to determine if the method of travel is sustainable based on a list of values
def is_sustainable(row):
    if row['Method of Travel'] in sustainable:
        return 'Sustainable'
    else:
        return 'Non-Sustainable'
    
# Adding a new column 'Sustainable?' by applying the function 'is_sustainable'
df['Sustainable?'] = df.apply(is_sustainable, axis=1)
df.head(5)
```

```python
# Reordering the columns to match the expected output
df_columns = ['Sustainable?', '% of trips per day', 'Trips per day', 'Number of Trips', 'Weekday', 'Method of Travel']
df = df[df_columns]

# Setting the column 'Sustainable?' as the index
df.set_index('Sustainable?', inplace=True)

# Exporting the DataFrame to a csv file
df.to_csv('Week 4 - Output.csv')
```
