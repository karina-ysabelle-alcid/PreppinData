```python
# import packages
import pandas as pd
from datetime import timedelta
```

```python
# import dataset
df = pd.read_csv('PD 2022 Wk 1 Input - Input.csv', index_col=0)
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'])
df.head(5)
```

```python
# Pupil's name
df['Pupil Name'] = df[["pupil first name", "pupil last name"]].agg(' '.join, axis=1)
df.head(5)
```

```python
# Function to get the student's birthday for the year 2022
def this_year(row):
    date = row['Date of Birth']
    return pd.to_datetime(f'2022/{date.month}/{date.day}') 

df["This Year's Birthday"] = df.apply(this_year, axis=1)
df.head(5)
```

```python
# Get day name from the student's birthday this year
df['Placeholder day_name'] = df["This Year's Birthday"].dt.day_name()
df['Placeholder Birthday'] = df["This Year's Birthday"]

# Function to shift weekend birthdays to friday
def to_friday(row):
    if row['Placeholder day_name'] == 'Saturday':
        row['Placeholder Birthday'] = row['Placeholder Birthday'] - timedelta(days=1)
    elif row['Placeholder day_name'] == 'Sunday':
        row['Placeholder Birthday'] = row['Placeholder Birthday'] - timedelta(days=2)

    # Fixes birthday's that reversed to 2021
    year = row['Placeholder Birthday'].year
    if year == 2021:
        row['Placeholder Birthday'].replace(year=2022)
    return row['Placeholder Birthday']

# Generating shifted birthday dates and its day name
df["Placeholder Birthday"] = df.apply(to_friday, axis=1)
df['Cake Needed On'] = df["Placeholder Birthday"].dt.day_name()
df.head(5)
```

```python
# Get month of birthday from the shifted date
df['Month'] = df["Placeholder Birthday"].dt.month_name()
df
```

```python
# Generating the number of birthdays held on a specific day each month using size()
df['BDs per Weekday and Month'] = df.groupby(['Month', 'Cake Needed On'],).transform('size')
df.head(5)
```

```python
# Functions to format the birthdays (DOB and this year's birthday) to DD/MM/YYYY 
def change_this_birthday(row):
    return row["This Year's Birthday"].strftime("%d/%m/%Y")
def change_DOB(row):
    return row["Date of Birth"].strftime("%d/%m/%Y")

# Formatting Birthday Columns
df["This Year's Birthday"] = df.apply(change_this_birthday, axis=1)
df["Date of Birth"] = df.apply(change_DOB, axis=1)
df.head(5)
```

```python
# Filtering to only include necessary columns
df_columns = ['Pupil Name', 'Date of Birth', "This Year's Birthday", 'Cake Needed On', 'Month', 'BDs per Weekday and Month']
df = df[df_columns]
df.head(5)
```

```python
# Setting Pupil Name as the index
df.set_index('Pupil Name',inplace=True)
df.to_csv('Week 2 - Output.csv')
```
