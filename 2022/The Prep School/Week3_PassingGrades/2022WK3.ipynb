```python
# import pandas
import pandas as pd
```

```python
# import students dataset
df_1 = pd.read_csv("PD 2022 Wk 1 Input - Input.csv")

# drop parental columns
df_1 = df_1.drop(df_1.iloc[:, 4:], axis=1)
df_1.head(5)
```

```python
# renaming 'gender' column to 'Gender'
df_1.rename(columns={'gender':'Gender'},inplace=True)
df_1.columns
```

```python
# import grades dataset
df_2 = pd.read_csv("PD 2022 WK 3 Grades.csv")
df_2.head(5)
```

```python
# Combine Students and Grades df
df = df_1.join(df_2)
df.head(5)
```

```python
# Get average score of each student
subjects = list(df_2.columns[1:])
df["Student's Avg Score"] = df[subjects].mean(axis=1).round(1)
df.head(5)
```

```python
# Get total number of passed subjects (>= 75)
df["Passed Subjects"] = df[subjects].ge(75).sum(axis=1)
df.head(5)
```

```python
# Filtering to only include necessary columns
df_columns = ['Passed Subjects', "Student's Avg Score", 'Student ID', 'Gender']
df = df[df_columns]
```

```python
# Setting Passed Subjects as the index
df.set_index('Passed Subjects', inplace=True)
df.to_csv('Week 3 - Output.csv')
```
