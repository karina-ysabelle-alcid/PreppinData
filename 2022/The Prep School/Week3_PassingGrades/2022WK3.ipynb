```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv("PD 2022 Wk 1 Input - Input.csv")
df_1 = df_1.drop(df_1.iloc[:, 4:], axis=1)
df_1.head(5)
```

```python
df_1.rename(columns={'gender':'Gender'},inplace=True)
df_1.columns
```

```python
df_2 = pd.read_csv("PD 2022 WK 3 Grades.csv")
df_2.head(5)
```

```python
df = df_1.join(df_2)
df.head(5)
```

```python
subjects = list(df_2.columns[1:])
df["Student's Avg Score"] = df[subjects].mean(axis=1).round(1)
df.head(5)
```

```python
df["Passed Subjects"] = df[subjects].ge(75).sum(axis=1)
df.head(5)
```

```python
df_columns = ['Passed Subjects', "Student's Avg Score", 'Student ID', 'Gender']
df = df[df_columns]
df.set_index('Passed Subjects', inplace=True)
df.to_csv('Week 3 - Output.csv')
```

```python
output_df = pd.read_csv("PD 2022 Wk 3 Output.csv", index_col=0)
output_df.sort_values('Student ID')
```

```python
def compare_dataframes(df1, df2):
    if df1.shape != df2.shape:
        print("Dataframes have different shapes.")
        return False
    
    if set(df1.columns) != set(df2.columns):
        print("Dataframes have different column names.")
        return False
    
    if df1.equals(df2):
        print("Dataframes are the same.")
        return True
    else:
        print("Dataframes are different.")
        return False

compare_dataframes(df, output_df)
```
